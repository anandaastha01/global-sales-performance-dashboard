```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set up visual style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
df = pd.read_csv("/content/global_sales_data.csv")
```

# Data cleaning and preprocessing

```python
print("shape of the dataset: ",df.shape)
```

```python
# check for any missing value
print("\n check if there is any null value \n :",df.isnull().sum())
```

```python
duplicate_records=df[df.duplicated()]
print("duplicate records in the dataset :",duplicate_records)
```

```python
# check for the number of duplicates
print("number of duplicated records :",df.duplicated().sum())
```

```python
# check for datatypes
print("\n check for the datatypes: \n",df.dtypes)
```

```python
df.head()
```

```python
# creating a backup data of the original dataset
df_copy=df.copy()
```

```python
df_copy.head()
```

```python
# Clean and standardize categorical columns
df['Segment'] = df['Segment'].str.strip().str.title()
df['Category'] = df['Category'].str.strip().str.title()
df['Ship Mode'] = df['Ship Mode'].replace({'Stnadard Class': 'Standard Class'})
df['Ship Mode'] = df['Ship Mode'].str.strip().str.title()
```

```python
# Convert numeric columns
numeric_columns = ['Sales', 'Profit', 'Unit Price']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Drop rows where key categorical values are missing
df = df.dropna(subset=['Order Date', 'Ship Date', 'Country', 'Region'])
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
# Step 6: Drop non-essential columns if present
df.drop(columns=['Returns', 'State', 'City'], axis=1, inplace=True, errors='ignore')
```

```python
# drop the duplicate records from data
df=df.drop_duplicates()
```

```python
df.to_csv("cleaned_global_sales.csv", index=False)
from google.colab import files
files.download("cleaned_global_sales.csv")
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Confirm the change
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
print(df.describe(include='all'))
```

```python
#  Sales by Country
country_sales = df.groupby("Country")["Sales"].sum().sort_values(ascending=False)
print("\nTop 5 Countries by Sales:\n", country_sales.head(5))
```

```python
# Plot: Top 05 countries by sales
country_sales.head(5).plot(kind='bar', color='skyblue', title='Top 5 Countries by Sales')
plt.ylabel("Total Sales")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sales by Product Category
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(f"sales by category",category_sales)
```

```python
df['Year'] = df['Order Date'].dt.year
# Group by year and sum sales
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()
print(yearly_sales)
```
